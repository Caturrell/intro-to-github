```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Mesh Set-up
#=======================================
N = 200
x_s, x_e = -4, 4
y_s, y_e = -2, 2

x = np.linspace(x_s, x_e, N)
y = np.linspace(y_s, y_e, N)

X, Y = np.meshgrid(x, y)
```

```python
# Function Defining
#=======================================
def uniform_velocity(U, aoa):
    u = U * np.ones( (N,N), dtype=float ) * np.cos(aoa)
    v = U * np.ones( (N,N), dtype=float ) * np.sin(aoa)
    return u, v
```

```python
# Parameters
#=======================================
U_0 = 1
alpha = (1/4) * math.pi

# Function Calling
#=======================================
u, v = uniform_velocity(U_0, alpha)
```

```python
# Streamline Plot
#=======================================
plt.figure(figsize=(10, 5))

plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.title("z-plane", fontsize=20)
plt.xlim(x_s, x_e)
plt.ylim(y_s, y_e)
plt.xticks([])
plt.yticks([])

# Streamline Plot
#=======================================
plt.streamplot(X, Y, u, v,
               density=0.75, linewidth=1.75, arrowsize=1.5, arrowstyle='-|>')
```
